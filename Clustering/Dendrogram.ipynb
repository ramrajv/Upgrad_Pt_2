```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/ramrajv/Upgrad_Pt_2/main/Clustering/Cricket.csv'
df = pd.read_csv(url, encoding='unicode_escape')
```

```python
df.columns
```

```python
df['Ave_Norm'] = (df['Ave'] - df['Ave'].mean())/df['Ave'].std()
```

```python
df['SR_Norm'] = (df['SR'] - df['SR'].mean())/df['SR'].std()
```

```python
import plotly.figure_factory as ff

X = df[['Ave_Norm','SR_Norm']]
labels = list(df['Player'].astype(str))
fig = ff.create_dendrogram(X, color_threshold=4, labels=labels, orientation = 'left')
fig.update_layout(width=800, height=1500)
fig.show()
```
